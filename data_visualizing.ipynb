## Download the data and take a quick look at dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

url = 'https://drive.google.com/uc?export=download&id=1ymlnauxZ1zoD0aboSWRnv_e5fxKMYLPA'
data = pd.read_csv(url)
```

```python
data.head(10)
```

```python
data.info()
```

## Try basic plotting using Matplotlib

### Analyse dependence between petal width and length

```python
data_copy = data.sort_values('petal_width')

plt.plot(data_copy['petal_width'], data_copy['petal_length'])
plt.title('Petal width to length dependence')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```

### Subplot and analyse dependence between sepal parameters and petal length

```python
data_copy = data.sort_values('petal_length')

plt.subplot(1, 2, 1)
plt.plot(data_copy['petal_length'].head(15), data_copy['sepal_width'].head(15), 'ms')
plt.title('Sepal width to petal length')

plt.subplot(1, 2, 2)
plt.plot(data_copy['petal_length'].head(15), data_copy['sepal_length'].head(15), 'r<')
plt.title('Sepal length to petal length')

plt.show()
```

### Check the mean sepal length depending on species

```python
species_mean = data.groupby('species').mean().reset_index()

species = species_mean['species']
sepal_length_means = species_mean['sepal_length']

plt.bar(species, sepal_length_means, color='green')
plt.title('Mean sepal length by species')
plt.xlabel('Species')
plt.ylabel('Mean sepal length')
plt.show()
```

## Try using another dataset and check its properties

```python
url_games = 'https://drive.google.com/uc?export=download&id=15FiZNK8ZMNQzpkG_bx056F3ezVFvi1lP'
data = pd.read_csv(url_games)

data.dropna(inplace=True)
data.head(10)
```

```python
data.info()
```

```python
data['User_Score'] = data['User_Score'].astype(float)
data.head(10)
```

## Using seaborn library, try to explore different visualisation methods

### Scatterplot of global sales for each year

```python
import seaborn as sns

sns.set_style('dark')
sns.scatterplot(data=data, x='Year_of_Release', y='Global_Sales')
```

### Lineplot of other sales for each year

```python
sns.lineplot(data=data, x='Year_of_Release', y='Other_Sales')
```

### Barplot of sales in Europe for every gaming platform

```python
platform_sales_max = data.groupby('Platform')['EU_Sales'].sum().reset_index()

sns.barplot(data=platform_sales_max, x='Platform', y='EU_Sales')
```

### Counts of every critic score and user score ever appeared as subplots

```python
plt.subplots(2, 1, figsize=(16, 12))
plt.subplot(2, 1, 1)
sns.countplot(data=data, x='Critic_Score')
plt.subplot(2, 1, 2)
sns.countplot(data=data, x='User_Score')
```

### Histogram for number of games appeared every 2.5 years, depending on rating gotten

```python
sns.histplot(data=data, x='Year_of_Release', binwidth=2.5, hue='Rating')
```

### Barplot of average user score for every genre introduced

```python
plt.figure(figsize=(14,5))
sns.barplot(data=data, x='Genre', y='User_Score', estimator=np.mean)
```

### Heatmap of how rating and genre impacted global sales

```python
pivot = data.pivot_table(
    index=['Genre'],
    columns=['Rating'],
    values='Global_Sales',
    aggfunc=np.mean
)

plt.figure(figsize=(15, 7))
sns.heatmap(pivot)
```
