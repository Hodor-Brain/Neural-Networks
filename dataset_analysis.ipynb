### Add some NaN's and then replace them with 0's

```python
import numpy as np

data_copy = data.copy()
data_copy.iloc[2:4, 5] = np.nan
data_copy.iloc[[2, 6, 9], 6] = np.nan
data_copy.iloc[5, 7] = np.nan

data_copy.head(10)
```

```python
data_copy.fillna(0, inplace=True)
data_copy.head(10)
```

## Delete all objects with NaN value

```python
data_copy = data.copy()
data_copy.iloc[2:4, 5] = np.nan
data_copy.iloc[[2, 6, 9], 6] = np.nan
data_copy.iloc[5, 7] = np.nan

data_copy.head(10)
```

## Create a new column with a total score for all exams

```python
data['total score'] = data['math score'] + data['reading score'] + data['writing score']
data.head(10)
```

## Check what gender did better on exams

```python
data.groupby('gender')['total score', 'math score', 'reading score', 'writing score'].mean()
```

### As a conclusion, we can say that female students showed better results than male ones on average.

## Check how lunch and test preparation impacted on results

```python
data.groupby(['lunch', 'test preparation course'])['total score'].mean().reset_index()
```

### As we see, better lunch as well as completed preparation resulted in better total score on average.
