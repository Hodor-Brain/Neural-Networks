## Download the data and take a quick look at dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

url = 'https://drive.google.com/uc?export=download&id=1KiOe3ae3mFb4vMOLUibh807kCDhZ5IUt'
data = pd.read_csv(url)
```

```python
data.head(10)
```

```python
data.info()
```

### Check transposed data

```python
first_three = data.head(3)
first_three.transpose()
```

### Delete columns that ends with '_lognorm'

```python
to_delete = [name for name in list(data.columns) if name.endswith('_lognorm')]
data.drop(to_delete, axis=1, inplace=True)
data
```

### Check the `describe` function

```python
data.describe().transpose()
```

### Make `'published'`column type a datetime

```python
from datetime import datetime

data['published'] = data['published'].astype('datetime64[ns]')
data.info()
```

```python
data.head(10)
```

### Create a few new columns based on `'published'`

```python
data['year'] = [d.year for d in data.published]
data['month'] = [d.month for d in data.published]
data['dayofweek'] = [d.isoweekday() for d in data.published]
data['hour'] = [d.hour for d in data.published]

data.head(10)
```

```python
data.info()
```

## Import seaborn for plotting

```python
import seaborn as sns

sns.set_style('darkgrid')
```

```python
year_month_data = data[((data['year'] == 2015) | (data['year'] == 2016)) & ((data['month'] == 3) | (data['month'] == 4))]
sns.countplot(data=year_month_data, x="month", hue="year")
```

As we see, March of 2015 had the highest number of publications among selected months

---

```python
march_2015_data = data[(data['year'] == 2015) & (data['month'] == 3)]
march_2015_data
```

```python
# plt.figure(figsize=(12,5))
sns.countplot(data=march_2015_data, x='dayofweek', hue='domain')
```

*   As we see, Tuesday had the highest number of publications on that month, while weekends had the lowest number
*   Habr had higher number of publications than geektimes on any day of week
*   On Saturday both domains had similar number of publications

```python
habrahabr = data[data["domain"] == "habrahabr.ru"]
habrahabr['daypart'] = np.where(habrahabr['hour'].between(6, 18, inclusive='left'), 'daytime', 'nighttime')
habrahabr
```

```python
sns.barplot(data=habrahabr, x="daypart", y="comments")
```

On Habr nighttime publications are receiving more comments than daytime ones
